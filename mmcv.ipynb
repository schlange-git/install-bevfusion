跑通bevfusion
首先建立新环境

conda create -n bev python=3.8 -y

conda activate bev

然后安装torch 

给出官网链接：https://pytorch.org/get-started/previous-versions/

运行：pip install torch==1.8.0+cu111 torchvision==0.9.0+cu111 torchaudio==0.8.0 -f https://download.pytorch.org/whl/torch_stable.html


然后需要安装mmcv：
需要注意，本环境需要的版本是 

mmcv 1.4.0  
mmdet  2.11.0  
mmdet3d 0.11.0  

这些安装包很多新版本和老版本已经不在同一个网站，直接点击切换版本会显示没有文档。
却换版本按钮为右下加绿色数字：


这里不建议使用mim安装各个版本的依赖，因为很容易无法solve environment。我这里给出pip 安装指令： 

pip install mmcv-full==1.4.0 -f https://download.openmmlab.com/mmcv/dist/cu111/torch1.8.0/index.html  

但是尽管如此，指令安装有时候仍然存在问题。因此可以尝试在spyder等ide中使用强制安装： 

!pip install mmcv-full==1.4.0 -f https://download.openmmlab.com/mmcv/dist/cu111/torch1.8.0/index.htm  
检查是否正确安装： 

#cmd  

pip show mmcv-full

#python 

import mmcv 

print(mmcv.__version__
)

如果存在pip得到结果，import无法完成可能是因为没有安装opencv  

运行： conda install -c conda-forge opencv 

建议使用conda安装，这里可以自动解决依赖问题，而且应该不会产生冲突。
